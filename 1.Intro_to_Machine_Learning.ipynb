Vamos a explicação da descrição de dados que o comando describe do pandas nos mostrou:

1.   count: Esta linha conta a quantidade de colunas não vazias (entraremos com o conceito de como os conjuntos de dados são encontrados no mundo real, rs)
2.   mean: Como sabemos, a média é a forma mais rápida de avaliarmos uma tendência central de um dataset e aqui ele está trazendo a média de cada coluna. 
3.   std: O desvio padrão nos mostra o quão distante os dados estão em relação a tendência central do conjunto. Ela é a raiz da variância, sendo útil porque a unidade de medida não sofre alteração igual na variância, facilitando a análise.
4.   min, 25%, 50%, 75%, max: 25%,50 %, e 75 % são quartis que dividem os dados ordenados em 4 partes iguais. Os valores mínimo e valor máximo destes dados ordenados são, respectivamente, os valores min e max.
